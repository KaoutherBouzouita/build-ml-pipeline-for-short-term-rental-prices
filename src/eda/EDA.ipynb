## Init a W&B run

```python
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
```

## Download the raw data from W&B

```python
local_path = wandb.use_artifact("sample.csv:latest").file()
```

## Convert Input Data to DataFrame

```python
df = pd.read_csv(local_path)
```

```python
df.head(50)
```

Some cells contain null values. Dates are in sring format.

### Observations:

There are 10 numeric features, 3 textual and 2 categorical.
Over 20000 missing cells. Like in last_review column.
Some variables are highly correlated.
Some outliers can be detected in longitude.

## Drop outliers

```python
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()
```

## Convert last_review to datetime

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```
